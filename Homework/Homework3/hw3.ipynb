```python
#1. load the dataset
from sklearn.datasets import load_breast_cancer
x, y = load_breast_cancer(return_X_y=True)
#x = breast_cancer.data
#y = breast_cancer.target
print(x)
print(y)


#2. split the dataset into training and test sets (70% data for the training set and 30% data for the test set).

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, train_size=0.70)
print(len(x_train))
print(len(x_test))





```

```python
#3. train a decison tree classifier by using the training set. 
#When you instanciate DecisionTreeClassifier, assign random_state=1 as a parameter of the class
#so that you can get deterministic decison tree (i.e., build the same tree).
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics

clf = DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)

y_pred = clf.predict(x_test)




#4. predict the test set's taret class (i.e., whether each patient will have a cancer or not), and then report accuracy

print("Decision Tree Classifier Accuracy:",metrics.accuracy_score(y_test, y_pred))



```

```python
#5. Draw/Visulaize the decison tree classifier's decison tree. You are welcome to use any python visualizaion class/library.

from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree

dataset = datasets.load_breast_cancer()

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=dataset.feature_names,  
                   class_names=dataset.target_names,
                   filled=True)





#6. repeat the same process for Random Forest Classifier 
#(i.e., training the Random Forest Classifier and then measure accuracy of the test set). Don't need to draw the Random Forest


from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=50)
classifier.fit(x_train,y_train)
y_pred_RandForest = classifier.predict(x_test)

print("Random Forest Accuracy: ", metrics.accuracy_score(y_test,y_pred_RandForest))



```
